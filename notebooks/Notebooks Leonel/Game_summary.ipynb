```python
import pandas as pd
ruta = "../../Data/Data Cruda/game_summary.csv"
df = pd.read_csv(ruta)
```

```python
df.head(100
        )
```

```python
# Ver cuántas filas duplicadas hay (excluyendo el índice)
duplicados = df.duplicated()
print("Cantidad de filas duplicadas:", duplicados.sum())
```

```python
# Mostrar las filas duplicadas
df_duplicadas = df[df.duplicated()]
print(df_duplicadas)
```

```python
# Mostrar todas las filas que están duplicadas (incluyendo los originales)
df_duplicados_completos = df[df.duplicated(keep=False)]

# Ver el resultado en forma de tabla
print(df_duplicados_completos)
```

```python
df_sin_duplicados = df.drop_duplicates().reset_index(drop=True)
df = df_sin_duplicados
```

```python
df.head(100)  # Mostrar las primeras 100 filas del DataFrame sin duplicado
```

```python
duplicados = df.duplicated()
print("Cantidad de filas duplicadas:", duplicados.sum())
```

```python
print(df.isnull().sum())
porcentaje_nulos = df.isnull().mean() * 100
print(porcentaje_nulos)
```

```python
print("Valores únicos no nulos de 'game_sequence':")
print(df['game_sequence'].dropna().unique()[:5])  # Muestra los primeros 5 distintos

print("\nValores únicos no nulos de 'game_status_text':")
print(df['game_status_text'].dropna().unique()[:5])
```

```python
df = df.drop(columns=['game_sequence', 'game_status_text','natl_tv_broadcaster_abbreviation','live_pc_time'])
```

```python
print(df.isnull().sum())
porcentaje_nulos = df.isnull().mean() * 100
print(porcentaje_nulos)
```

```python
# Reemplazá la ruta con la que quieras
ruta_guardado = "../../Data/Data Limpia/game_summary_limpio.csv"
# Guardar sin el índice
df.to_csv(ruta_guardado, index=False)
```
