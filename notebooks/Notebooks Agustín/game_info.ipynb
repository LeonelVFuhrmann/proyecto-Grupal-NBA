```python
# üìå Importamos las librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# üìå Ruta del archivo CSV #game_info.csv
ruta = "../../Data/Data Cruda/game_info.csv"

# üìå Cargamos el CSV
df_inactive = pd.read_csv(ruta)

# üìå Mostramos estructura general del DataFrame
df_inactive.info()
df_inactive.head()
df_inactive.shape
```

```python
# üìå Mostrar cantidad de valores nulos por columna
nulos = df_inactive.isnull().sum()
nulos = nulos[nulos > 0].sort_values(ascending=False)

print("üìä Valores nulos por columna:")
print(nulos)
```

```python
# üìå Mostrar filas que contienen al menos un valor nulo
filas_con_nulos = df_inactive[df_inactive.isnull().any(axis=1)]

# üìå Ver las primeras filas con nulos
filas_con_nulos.head()
```

```python
# üìå Eliminar la fila que contiene valores nulos en attendance o game_time
df_inactive = df_inactive[~(df_inactive['attendance'].isnull() | df_inactive['game_time'].isnull())]
```

```python
# üìå Mostrar filas que contienen al menos un valor nulo
filas_con_nulos = df_inactive[df_inactive.isnull().any(axis=1)]

# üìå Ver las primeras filas con nulos
filas_con_nulos.head()
```

```python
# üìå Verificar si hay filas duplicadas en todo el DataFrame
duplicados = df_inactive.duplicated()

# üìå Mostrar cu√°ntas filas est√°n duplicadas
print(f"Filas duplicadas: {duplicados.sum()}")

# üìå Mostrar las filas duplicadas
df_inactive[duplicados]
```

```python
# üìå Paso 1: Detectar filas duplicadas (todas las columnas)
duplicados = df_inactive[df_inactive.duplicated(keep=False)]

# üìå Paso 2: Contar cu√°ntas veces se repite cada fila duplicada
resumen_duplicados = duplicados.value_counts().reset_index(name='Repeticiones')

# üìå Paso 3: Mostrar resumen
print(f"üîÅ Total de filas duplicadas: {len(duplicados)}")
print("üìã Resumen de duplicados (filas id√©nticas):")
print(resumen_duplicados.head())

# üìå Paso 4: Mostrar tabla de duplicados
duplicados
```

```python
# üìå Paso 1: Detectar duplicados en la columna 'game_id'
duplicados_id = df_inactive[df_inactive.duplicated(subset='game_id', keep=False)]

# üìå Paso 2: Mostrar los duplicados encontrados
print(f"üîÅ Total de filas con id duplicado: {len(duplicados_id)}")
duplicados_id.head()
```

```python
df_inactive = df_inactive[~df_inactive.duplicated(subset='game_id', keep=False)]
```

```python
# ‚úîÔ∏è Convertir a datetime
df_inactive['game_date'] = pd.to_datetime(df_inactive['game_date'], errors='coerce')

# ‚úîÔ∏è Filtrar por fecha >= 2018-01-01
df_inactive = df_inactive[df_inactive['game_date'] >= '2018-01-01']
```

```python
# üìå Exportar el DataFrame limpio a CSV en la carpeta del proyecto
ruta_salida = "../../Data/Data Limpia/game_info.csv"
df_inactive.to_csv(ruta_salida, sep=',', index=False, decimal='.')

print("‚úÖ Archivo exportado correctamente como game_info.csv")
```
