LIMPIEZA DE CSV LINE_SCORE

```python
# üìå Importamos las librer√≠as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üìå Ruta del archivo CSV
ruta = "../../Data/Data Limpia/line_score.csv"

# üìå Cargamos el CSV
df_line_score = pd.read_csv(ruta)

```

```python
# Dimensiones
print("Dimensiones:", df_line_score.shape)
```

```python
# Tipos de datos
print("Tipos de datos:")
print(df_line_score.dtypes)
```

```python
# Columnas
print("Columnas:")
print(df_line_score.columns.tolist())
```

```python
# Valores nulos por columna
nulos = (df_line_score.isnull().sum())
total_filas = len(df_line_score)
porcentaje = (nulos / total_filas * 100).round(2)

# Mostrar ambos
print("Valores nulos por columna (cantidad y porcentaje):\n")
for col in df_line_score.columns:
    print(f"{col}: {nulos[col]} nulos ({porcentaje[col]}%)")
```

```python
# Verificar filas duplicadas
duplicados = df_line_score[df_line_score.duplicated()]
print(f"Cantidad de filas duplicadas: {duplicados.shape[0]}")

# Mostrar duplicados si existen
if not duplicados.empty:
    print(duplicados)
```

- CONVERTIMOS LA COLUMNA DE GAME_DATE_EST A TIPO DATETIME

```python
# Asegurarse de que la columna existe
if 'game_date_est' in df_line_score.columns:
    df_line_score['game_date_est'] = pd.to_datetime(df_line_score['game_date_est'], errors='coerce')
    print("‚úÖ 'game_date_est' convertida a formato fecha.")
else:
    print("‚ùå La columna 'game_date_est' no existe.")
```

- CAMBIAMOS LAS COLUMNAS FLOAT64 A INT64

```python
# Seleccionar columnas tipo float64
float_cols = df_line_score.select_dtypes(include=['float64']).columns

# Intentar convertir a int64 (solo si no hay decimales ni nulos)
for col in float_cols:
    if df_line_score[col].isnull().sum() == 0 and (df_line_score[col] % 1 == 0).all():
        df_line_score[col] = df_line_score[col].astype('int64')
        print(f"‚úÖ Columna '{col}' convertida a int64.")
    else:
        print(f"‚ö†Ô∏è Columna '{col}' no se pudo convertir (tiene nulos o decimales).")
```

- VERIFICO DE QUE ANOS SON LAS FECHAS EN LA COLUMNA GAME_DATE_EST

```python
# Verificar el rango de a√±os en la columna game_date_est
min_year = df_line_score['game_date_est'].min().year
max_year = df_line_score['game_date_est'].max().year

print(f"üìÖ Rango de a√±os: {min_year} - {max_year}")
```

- RELLENAMOS LOS NULOS CON EL NUMERO 0 EN LAS SIGUIENTES COLUMNAS pts_qtr1_home hasta pts_ot10_home inclusive Y pts_qtr1_away hasta pts_ot10_away inclusive Y ADEMAS game_sequence


```python
# Rellenar valores nulos con 0 en columnas de puntos por cuarto y overtime (home)
cols_home = df_line_score.loc[:, 'pts_qtr1_home':'pts_ot10_home'].columns
df_line_score[cols_home] = df_line_score[cols_home].fillna(0)

# Rellenar valores nulos con 0 en columnas de puntos por cuarto y overtime (away)
cols_away = df_line_score.loc[:, 'pts_qtr1_away':'pts_ot10_away'].columns
df_line_score[cols_away] = df_line_score[cols_away].fillna(0)

# Rellenar valores nulos en la columna game_sequence
df_line_score['game_sequence'] = df_line_score['game_sequence'].fillna(0)
```

```python
df_line_score
```

- FILTRAMOS POR FECHA >2018 INCLUSIVE

```python
# Asegurarse de que la columna est√© en formato datetime
df_line_score['game_date_est'] = pd.to_datetime(df_line_score['game_date_est'], errors='coerce')

# Filtrar por fechas desde el 1 de enero de 2018 (inclusive)
df_line_score_mayor2018 = df_line_score[df_line_score['game_date_est'] >= '2018-01-01']
```

```python
duplicados = df_line_score_mayor2018.duplicated(subset='game_id').sum()
print(f"üîÅ Filas duplicadas por game_id: {duplicados}")

# Mostrar todas las filas que tienen game_id duplicado (incluye la primera aparici√≥n)
duplicados_game_id = df_line_score_mayor2018[df_line_score_mayor2018.duplicated(subset='game_id', keep=False)]

# Mostrar el resultado
print(duplicados_game_id)
```

- ELIMINAMOS DUPLICADOS DE LA COLUMNA GAME_ID

```python
df_line_score_mayor2018 = df_line_score_mayor2018.drop_duplicates(subset='game_id', keep='first')
```

- VERIFICAMOS CUANTAS FILAS QUEDAN DESPUES DE ELIMINAR DUPLICADOS Y CUANTOS DUPLICADOS HAY EN GAME_ID

```python
print(f"N√∫mero de filas despu√©s de eliminar duplicados: {len(df_line_score_mayor2018)}")
print(f"Duplicados restantes en game_id: {df_line_score_mayor2018.duplicated(subset='game_id').sum()}")
```
